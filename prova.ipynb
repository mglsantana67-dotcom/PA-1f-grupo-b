```python
idade = int(input("digite idade ."))

if idade <= 17:
  categoria = "menor de idade"
elif idade <= 60:
  categoria = "maior de idade"
else:
  categoria = "idoso"

print(f"você tem {idade} anos e é um {categoria}.")
```

```python

valor_compra = float(input("Digite o valor da compra: R$"))

disconto = 0
valor_final = 0

if valor_compra <= 100:
  desconto = 0
elif valor_compra <= 500:
  desconto = 0.10
else: # Acima de R$500
  desconto = 0.20

valor_final = valor_compra * (1 - desconto)

print(f"O valor final da compra com desconto é: R${valor_final:.2f}")
```

```python

numero1 = int(input("digite o primeiro lado: "))
numero2 = int(input("digite o segundo lado: "))
numero3 = int(input("digite o terceiro lado: "))

if (numero1 + numero2 > numero3) and (numero1 + numero3 > numero2) and (numero2 + numero3 > numero1):
    print("Os valores informados podem formar um triângulo válido!")
else:
    print("Os valores informados NÃO podem formar um triângulo.")
```

```python
energia = int(input("Digite a energia do personagem (0-100): "))

if energia == 0:
    status = "Morto"
elif energia <= 30:
    status = "Crítico"
elif energia <= 70:
    status = "Normal"
elif energia <= 100:
    status = "Saudável"
else:
    status = "Energia inválida (acima de 100)"

print(f"Energia: {energia} - Status: {status}")
```

```python
media = float(input("Digite a média final do aluno: "))

if media >= 7:
    resultado = "Aprovado"
elif media >= 5:
    resultado = "Recuperação"
else:
    resultado = "Reprovado"

print(f"Média: {media:.1f} - Status: {resultado}")
```

```python
salario = float(input("Salário: R$"))
pontuacao = int(input("Pontuação (0-10): "))

if pontuacao < 5:    pct = 0
elif pontuacao <= 7: pct = 0.05
elif pontuacao <= 9: pct = 0.10
else:                pct = 0.15

bonus = salario * pct

print(f"Bonus ({pct*100}%): R${bonus}")
print(f"Total: R${salario + bonus}")
```
