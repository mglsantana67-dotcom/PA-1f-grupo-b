```python
# Escrever um programa q o usuario digita um numero de 1 a 10  e o programa imprimi a tabuada do numero digitado

numero = int(input("Digite um número de 1 a 10: "))
print(f"Tabuada do {numero}:")
for i in range(1, 11):
  print(f"{numero} X {i} = {numero * i}")
```
