```python
nota = float(input("Digite a nota final do aluno: "))

if nota >= 7:
    resultado = "Aprovado"
elif nota >= 5:
    resultado = "Recuperação"
else:
    resultado = "Reprovado"

print(f"Média: {nota:.1f} - Status: {resultado}")
```

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  categoria = "criança"
elif idade <= 17:
  categoria = "adolescente"
elif idade <= 60:
  categoria = "adulto"
else:
  categoria = "idoso"

print(f"você tem {idade} anos e é um {categoria}.")
```

```python
dia_numero = int(input("Digite um número de 1 a 7: "))

if dia_numero == 1:
    dia = "Domingo"
elif dia_numero == 2:
    dia = "Segunda-feira"
elif dia_numero == 3:
    dia = "Terça-feira"
elif dia_numero == 4:
    dia = "Quarta-feira"
elif dia_numero == 5:
    dia = "Quinta-feira"
elif dia_numero == 6:
    dia = "Sexta-feira"
elif dia_numero == 7:
    dia = "Sábado"
else:
    dia = "Número inválido! Digite um valor entre 1 e 7."

print(f"O dia correspondente é: {dia}")
```
